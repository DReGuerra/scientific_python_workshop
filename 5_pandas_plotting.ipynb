## Read data files
### .xlsx files
We use the attribute <b>read_excel()</b> in the pandas module. The local variable assigned in the import, <b>pd</b>, is used to call the attribute. The target file name is used as an input parameter.

```python
covid19_file = "1_data\covid_19_data.xlsx"
covid_DF = pd.read_excel(covid19_file)
```

## Pandas plotting functionality
Pandas has some plotting functionality from an underlying package (matplotlib). \
Now, we can plot a simple histogram on the sum of `Deaths` that our dataframe reports in Australia as grouped by `Province/State`.

```python
covid_DF[covid_DF['Country/Region'] == 'Australia'].groupby('Province/State')['Deaths'].sum().plot(kind='bar').set_ylabel("Totals")
```

We can plot a simple histogram on the sum of cases of `Confirmed`, `Deaths`, and `Recovered` cases that our dataframe reports in Australia as grouped by `Province/State`.

```python
covid_DF[covid_DF['Country/Region'] == 'Australia'].groupby('Province/State')[['Confirmed','Deaths','Recovered']].sum().plot(kind='bar',stacked=False,title='Australia totals of confirmed, deaths, and recovered cases')
```

We can plot a simple histogram on the sum of cases of `Confirmed`, `Deaths`, and `Recovered` cases that our dataframe reports in Germany as grouped by `Province/State`.

```python
covid_DF[covid_DF['Country/Region'] == 'Italy'].groupby('Province/State')[['Confirmed','Deaths','Recovered']].sum().plot(kind='bar',stacked=False,title='Italy totals of cases')
```
